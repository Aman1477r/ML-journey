```python
df_copy = df.copy()
```

```python
df_copy.drop(df_copy.index[10472], inplace = True)
```

```python
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('M', '000')
df_copy['Size'] = df_copy['Size'].str.replace('k', '')
df_copy['Size'] = df_copy['Size'].replace('Varies with device', np.nan)
df_copy['Size'] = df_copy['Size'].astype(float)
```

```python
df_copy['Size'].unique()
```

```python
df['Installs'].unique()
```

```python
char_remove = ['+',',','$']
cols_to_clean = ['Installs','Price']
for item in char_remove:
    for col in cols_to_clean:
        df_copy[col] = df_copy[col].str.replace(item,'')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'] = df_copy['Installs'].astype(int)
df_copy['Price'] = df_copy['Price'].astype(float)
```

```python
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])
```

```python
df_copy['Day'] = df_copy['Last Updated'].dt.day
df_copy['Month'] = df_copy['Last Updated'].dt.month
df_copy['Year'] = df_copy['Last Updated'].dt.year
```

```python
df_copy.head()
```

```python
df_copy['Android Ver'].unique()
```

```python
df_copy = df_copy[df_copy['Android Ver'] != '1.0.19']
df_copy = df_copy[df_copy['Android Ver'] != '19.0.0']
```

```python
df_copy['Android Ver'].unique()
```

```python
df_copy.to_csv('googleplaystore_cleaned.csv', index = False)
```

```python
# Total installs by category
category_installs = df.groupby('Category')['Installs'].sum()

# Sort in descending order
category_installs = category_installs.sort_values(ascending=False)

print(category_installs)
```

```python
top_category = df.groupby('Category')['Installs'].sum().idxmax()
top_installs = df.groupby('Category')['Installs'].sum().max()

print("Category:", top_category)
print("Total Installs:", top_installs)
```

```python
import matplotlib.pyplot as plt

category_installs = df_copy.groupby('Category')['Installs'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
category_installs.plot(kind='bar')
plt.title('Total Installs by Category')
plt.xlabel('Category')
plt.ylabel('Total Installs')
plt.show()
```

```python
popular_categories = category_installs.head(10).index.tolist()

all_top_apps_data = []

for category in popular_categories:
    print(f"\n=== {category} ===")

    top5_category = (
        df_copy[df_copy['Category'] == category]
        .nlargest(5, 'Installs')[['App', 'Installs']]
    )
    top5_category['Category'] = category # Add Category column for hue in plot
    all_top_apps_data.append(top5_category)

    print(top5_category.drop(columns=['Category']).to_string(index=False))

top_apps = pd.concat(all_top_apps_data, ignore_index=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 8))

sns.barplot(
    data=top_apps,
    x='Installs',
    y='App',
    hue='Category'
)

plt.title('Top Installed Apps by Category')
plt.tight_layout()
plt.show()
```
