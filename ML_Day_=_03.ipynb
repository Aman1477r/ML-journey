1. ## One Hot Encoding


One-Hot Encoding converts each category into a separate binary column containing 0s and 1s.


Nominal data (categories with no order)
Features with a small number of unique values
Advantage

```python
df = pd.DataFrame({
    'color':['red','blue','green','green','red','blue']

})
```

```python
encoder = OneHotEncoder()
```

```python
encodee= encoder.fit_transform(df[['color']]).toarray()
```

```python
dff = pd.DataFrame(encodee,columns=encoder.get_feature_names_out())
```

2 ### Ordinal encoding
Ordinal Encoding assigns numerical values to categories based on their natural order or ranking.

Example
Size
Small
Medium
Large


```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
sdd = pd.DataFrame({
    'size':['small','medium','large','medium','small','large']
})
```

```python
sdd
```

```python
ordinal_en = OrdinalEncoder()
```

```python
ordinal_en.fit_transform(sdd)
```

```python
ordinal_en.transform([['small']])
```

#### TARGEEEEET Encoding


Target Encoding replaces each category with the average value of the target variable for that category.

Example
City	Purchased
Mumbai	1
Mumbai	0
Delhi	1

```python
import pandas as pd
```

```python
gg = pd.DataFrame({
    'city': ['New York','Los Angeles','Chicago','New York','Chicago'],
    'price': [200,400,500,600,700]
})
```

```python
gg
```

```python
mean_price =  gg.groupby('city')['price'].mean()
```

```python
gg['encoded_data'] = gg['city'].map(mean_price)
```

```python
gg
```

```python
gg[['price','encoded_data']]
```
