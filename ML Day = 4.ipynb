```python
df['firstname'] = df['flight'].str.split('-').str[0]
df['lastname'] = df['flight'].str.split('-').str[1]
```

```python
df.head()
```

```python
df.info()
```

```python
df['firstname'] = df['firstname'].astype(str)
df['lastname'] = df['lastname'].astype(int)
```

```python
df.info()
```

```python
df['stops'].unique()
```

```python
df.isnull().sum()
```

```python
df['stops'] = df['stops'].map({
    'zero': 0,
    'one': 1,
    'two_or_more': 2
})
```

```python
df.head(2)
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df['duration_hour'] = df['duration'].astype(str).str.split('.').str[0].astype(int)
df['duration_min'] = df['duration'].astype(str).str.split('.').str[1].astype(int)
```
